## Text to Handwriting using Python

```python
import pywhatkit as kit
import cv2
```

```python
your_input= input("Type the text you want to make it Handwriting: \n\n")

kit.text_to_handwriting(your_input, save_to="handwriting.png")
img = cv2.imread("handwriting.png")
cv2.imshow("Text to Handwriting", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
